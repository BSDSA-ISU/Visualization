# Educational Resource Allocation Dataset

[Dataset Code Link](https://www.kaggle.com/datasets/colabsss/educational-resource-allocation-dataset)

## About this dataset
This dataset contains 5700 institutional-level indicators related to the distribution and utilization of educational resources. It focuses on factors that influence how effectively institutions manage teaching capacity, infrastructure availability, technology access, and academic engagement. The dataset captures variations in institutional characteristics such as student population, teaching staff availability, classroom infrastructure, and digital learning resources. The purpose of this dataset is to provide a structured representation of educational resource distribution across institutions. It includes quantitative measures describing classroom capacity, laboratory availability, digital learning tools, infrastructure quality, operational expenditure, and resource utilization levels. These indicators help in understanding how educational institutions balance academic demand with available infrastructure and human resources. Institutional characteristics such as student–teacher ratios, classroom utilization levels, and technology availability provide insights into the learning environment within each institution. Additional indicators related to learning outcomes, student engagement, and infrastructure quality help represent the overall effectiveness of institutional resources. The dataset also includes information about operational costs and resource wastage indicators, which reflect how efficiently educational resources are managed. By combining academic, infrastructural, and utilization-related indicators, the dataset offers a comprehensive perspective on institutional resource allocation.

```python
print("Mean:",df["Total_Students"].mean())
print("Median:", df["Total_Students"].median())
print("Mode:", df["Total_Students"].mode())
print("Standard Deviation:", df["Total_Students"].std())
print("Min value:", df["Total_Students"].min())
print("Max value:", df["Total_Students"].max())
```
